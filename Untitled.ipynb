```python
irrelevant = ['match_id', 'season', 'start_date', 'non_striker','extras', 'wides', 'noballs', 'byes', 'legbyes',
       'penalty', 'wicket_type', 'player_dismissed', 'other_wicket_type',
       'other_player_dismissed']
print(f'Before Removing Irrelevant Columns : {ipl_data.shape}')
ipl_data = ipl_data.drop(irrelevant, axis=1) # Drop Irrelevant Columns
print(f'After Removing Irrelevant Columns : {ipl_data.shape}')
ipl_data.head()
```

```python
ipl_data['batting_team'].unique()
```

```python
ipl_data.loc[0, 'Total_runs'] = ipl_data.loc[0, 'runs_off_bat']
ipl_data.head()
```

```python
for i in range(1, len(ipl_data)):
        ipl_data.loc[i, 'Total_runs'] =ipl_data.loc[i-1, 'Total_runs']+ipl_data.loc[i, 'runs_off_bat']

#ipl_data['Total_runs'] = ipl_data.runs_off_bat + ipl_data.Total_runs.shift(1)
ipl_data.head(50)
```

```python
const_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
              'Mumbai Indians', 'Kings XI Punjab', 'Royal Challengers Bangalore',
              'Delhi Daredevils', 'Sunrisers Hyderabad']
```

```python
print(f'Before Removing Inconsistent Teams : {ipl_data.shape}')
ipl_data = ipl_data[(ipl_data['batting_team'].isin(const_teams)) & (ipl_data['bowling_team'].isin(const_teams))]
print(f'After Removing Irrelevant Columns : {ipl_data.shape}')
print(f"Consistent Teams : \n{ipl_data['batting_team'].unique()}")
ipl_data.head()
```

```python
ipl_data['venue'].unique()
```

```python
const_venues = ['Wankhede Stadium','M Chinnaswamy Stadium','Eden Gardens','Arun Jaitley Stadium','Sardar Patel Stadium']
```

```python
print(f'Before Removing Inconsistent venues : {ipl_data.shape}')
ipl_data = ipl_data[(ipl_data['venue'].isin(const_venues))]
print(f'After Removing Irrelevant Columns : {ipl_data.shape}')
print(f"Consistent venues : \n{ipl_data['venue'].unique()}")
ipl_data.head()
```

```python
ipl_data = ipl_data.rename(columns = {"striker":"batsmen"})
ipl_data.head()

```

```python
print(f'Before Removing Overs : {ipl_data.shape}')
ipl_data = ipl_data[ipl_data['ball'] >= 5.0]
print(f'After Removing Overs : {ipl_data.shape}')
ipl_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
for col in ['bowler','venue']:
    ipl_data[col] = le.fit_transform(ipl_data[col])
ipl_data.head()
```

```python
#df.head()
ipl_data = pd.get_dummies(data=ipl_data, columns=['batting_team', 'bowling_team', 'venue', 'batsmen','bowler'])
ipl_data.head()
```

```python
ipl_data.columns
```

```python
ipl_data = ipl_data.astype(float)
```

```python
ipl_data.dtypes
```

```python
ipl_data.head(15)
```

```python
features = ipl_data.drop(['Total_runs'], axis=1)
labels = ipl_data['Total_runs']
```

```python
# Perform 80 : 20 Train-Test split
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.20, shuffle=True)
print(f"Training Set : {train_features.shape}\nTesting Set : {test_features.shape}")
```

```python
models = dict()
```

1. Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor()
# Train Model
tree.fit(train_features, train_labels)
```

```python
# Evaluate Model
train_score_tree = str(tree.score(train_features, train_labels) * 100)
test_score_tree = str(tree.score(test_features, test_labels) * 100)
print(f'Train Score : {train_score_tree[:5]}%\nTest Score : {test_score_tree[:5]}%')
models["tree"] = test_score_tree
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
print("---- Decision Tree Regressor - Model Evaluation ----")
print("Mean Absolute Error (MAE): {}".format(mae(test_labels, tree.predict(test_features))))
print("Mean Squared Error (MSE): {}".format(mse(test_labels, tree.predict(test_features))))
print("Root Mean Squared Error (RMSE): {}".format(np.sqrt(mse(test_labels, tree.predict(test_features)))))
```

```python
def predict_score( innings, batting_team, bowling_team, batsmen, bowler, model=svm):
    prediction_array = []
  
    prediction_array =  [ innings,batting_team, bowling_team, batsmen, bowler]
    prediction_array = np.array([prediction_array])
    pred = model.predict(prediction_array)
    return int(round(pred[0]))
```
